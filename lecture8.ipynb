## Try and Except

```python
try:
    a= int(input("Enter a number"))
    b = int(input("Enter a number"))
    c= a/b
except ZeroDivisionError:
    print("you have entered denominator as zero")
except ValueError:
    print("you are converting text into int")
except Exception:
    print("there is somthing wrong that i'm unable to find")
else:
     print(c)
finally:
    print("succefully code completed")
```
