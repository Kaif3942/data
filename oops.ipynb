```python
class Car():
    name=""
    def __init__(self,_name):
        self.name = _name
    def car_detail(self):
        print("hey, this is my {0}".format(self.name))

class Toyota(Car):
    def __init__(self,name,model):
        Car.__init__(self,name)
        self.model = model

class Fortuner(Toyota):
    def __init__(self,name,model,price):
        Toyota.__init__(self,name,model)
        self.price = price


toyota= Toyota("toyota","fortuner")
toyota.car_detail()

maruti = Car("maruti")
maruti.car_detail()
```

```python
from abc import ABC,abstractmethod
class Bike(ABC): #abstract class
    @abstractmethod #abstract method
    def gear(self):
        pass
class KTM(Bike):
    def gear(self):
        return 5
ktm = KTM()
print(ktm.gear())
```
